```python
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# Define input and output for logic gates
# For NOT gate, only one input is needed
# For AND and OR gates, two inputs are needed
X_and = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_and = np.array([[0], [0], [0], [1]])

X_or = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y_or = np.array([[0], [1], [1], [1]])

X_not = np.array([[0], [1]])
y_not = np.array([[1], [0]])
```

```python
# Define neural network architecture for AND & OR
def create_model():
    model = Sequential()
    model.add(Dense(1, input_dim=X_and.shape[1], activation='sigmoid'))
    model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

**model.add()**: This method is used to add layers to the neural network model. In this case, it's adding a Dense layer.

**Dense(1, input_dim=X_and.shape[1], activation='sigmoid'):**

**Dense(1, ...):** This part initializes a fully connected (dense) layer with 1 neuron. The number 1 indicates the number of neurons in the layer. For this simple example, since we are implementing basic logic gates (AND, OR, NOT), we only need one neuron to represent the output.

**input_dim=X_and.shape[1]:** This specifies the number of inputs to the layer.

**X_and.shape[1]** gives the number of columns in the input data X_and, which corresponds to the number of input features. This parameter is necessary for the first layer of the model to specify the input shape.

**activation='sigmoid':** This sets the activation function for the layer. In this case, it's a sigmoid activation function. The sigmoid function squashes the output of each neuron to the range [0, 1], making it suitable for binary classification tasks.

**Sequential** is used to define the architecture of the model by stacking layers sequentially, while **Dense** is used to define the individual layers within the model, specifically for fully connected layers.
**Sequential** can contain various types of layers, but **Dense** is specifically for fully connected layers where all neurons are connected to each neuron in the previous and next layers.

```python
# Train AND gate model
model_and = create_model()
model_and.fit(X_and, y_and, epochs=1000, verbose=0)

# Train OR gate model
model_or = create_model()
model_or.fit(X_or, y_or, epochs=1000, verbose=0)
```

```python
# Define neural network architecture
model_not = Sequential()
model_not.add(Dense(1, input_dim=1, activation='sigmoid'))
model_not.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Train NOT gate model
model_not.fit(X_not, y_not, epochs=1000, verbose=0)
```

```python
# Test the trained models
test_cases = [[0, 0], [0, 1], [1, 0], [1, 1]]
test_cases_NOT = [[0], [1]]
```

```python
print("AND Gate Predictions:")
for test_case in test_cases:
    prediction = model_and.predict(np.array([test_case]))
    print(test_case, '-->', round(prediction[0][0]))
```

```python
print("\nOR Gate Predictions:")
for test_case in test_cases:
    prediction = model_or.predict(np.array([test_case]))
    print(test_case, '-->', round(prediction[0][0]))
```

```python
print("\nNOT Gate Predictions:")
for test_case in test_cases_NOT:
    prediction = model_not.predict(np.array([test_case]))
    print(test_case, '-->', round(prediction[0][0]))
```

```python
training_data = np.array([[0,0],[0,1],[1,0],[1,1]], "float32")
target_data = np.array([[0],[1],[1],[0]], "float32")

model = Sequential()
model.add(Dense(16, input_dim=2, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
model.compile(loss='mean_squared_error',
                       optimizer='adam',
                       metrics=['binary_accuracy'])
model.fit(training_data, target_data, epochs=1000)
scores = model.evaluate(training_data, target_data)

print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
print (model.predict(training_data).round())
```
